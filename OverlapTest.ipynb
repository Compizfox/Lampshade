```python
%matplotlib inline
%config InlineBackend.figure_format ='svg'
import numpy as np
import matplotlib.pyplot as plt

from sys import path
path.append("../../shared_python_scripts/")

from BrushDensityParser import BrushDensityParser

bdp = BrushDensityParser()

directory = '/home/lars2/Nextcloud/Eos/Data/School/CE/Masteropdracht/MD SPiVHB/2019-06-26 4x4_grid/'

densPoly = bdp.loadDensity(directory + 'grid_1.0_1.0/PolyDens.dat')
densSolv = bdp.loadDensity(directory + 'grid_1.0_1.0/SolvDens.dat')
poly_ta = np.mean(densPoly, axis=0)[:, 2]
solv_ta = np.mean(densSolv, axis=0)[:, 2]

fig, ax = plt.subplots()
ax.plot(poly_ta)
ax.plot(solv_ta)
ax.plot(solv_ta * poly_ta)
fig.show()

np.trapz(solv_ta * poly_ta)
```
